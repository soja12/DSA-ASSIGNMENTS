# **Case Study on Testing of Hypothesis**

2. A company started to invest in digital marketing as a new
way of their product promotions. For that they collected
data and decided to carry out a study on it

a. The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('/content/Sales_add.csv')
df.head()
```

```python
df.info()
```

```python
sales_before = df['Sales_before_digital_add(in $)'].sum()
sales_before
```

```python
sales_after = df['Sales_After_digital_add(in $)'].sum()
sales_after
```

```python
if sales_after > sales_before:
    print("There is an increase in sales after stepping into digital marketing.")
elif sales_after < sales_before:
    print("There is a decrease in sales after stepping into digital marketing.")
else:
    print("There is no change in sales after stepping into digital marketing.")
```

b. The company needs to check whether there is any
dependency between the features “Region” and
“Manager”.

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('/content/Sales_add.csv')
df.head()
```

```python
contingency_table = pd.crosstab(df['Region'],df['Manager'])
contingency_table
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

```python
alpha = 0.05
```

```python
if p < alpha:
    print("There is a significant dependency between Region and Manager.")
else:
    print("There is no significant dependency between Region and Manager.")
```
