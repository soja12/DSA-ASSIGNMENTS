```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

a. Find out how many males and females participated in
the test.

```python
gender_counts = df['gender'].value_counts()
gender_counts
```

b. What do you think about the students' parental level of
education?

```python
parental_education_counts = df['parental level of education'].value_counts()
parental_education_counts
```

```python
mean_scores_by_education_level = df.groupby('parental level of education').mean()
mean_scores_by_education_level
```

c. Who scores the most on average for math, reading and
writing based on
i. Gender
ii. Test preparation course

```python
avg_scores = df.groupby('gender').mean()
avg_scores
```

```python
avg_scores_course = df.groupby('test preparation course').mean()
avg_scores_course
```

d. What do you think about the scoring variation for
math, reading and writing based on
i. Gender
ii. Test preparation course

```python
score_variation = df.groupby('gender').std()
score_variation
```

```python
prep_course = df.groupby('test preparation course').std()
prep_course
```

e. The management needs your help to give bonus points
to the top 25% of students based on their math score,
so how will you help the management to achieve this.

```python
cutoff_score = df['math score'].quantile(0.75)
cutoff_score
```

```python
top_students = df[df['math score'] >= cutoff_score]
top_students
```

```python
bonus_points = 5
```

```python
top_students['math score'] += bonus_points
top_students.head()
```
