```python
import pandas as pd
```

```python
df = pd.read_csv("/content/beer-servings.csv")
df.head()
```

```python
Q1 = df['beer_servings'].quantile(0.25)
Q3 = df['beer_servings'].quantile(0.75)
IQR = Q3 - Q1
```

```python
df_iqr = df[(df['beer_servings'] >= Q1 - 1.5 * IQR) & (df['beer_servings'] <= Q3 + 1.5 * IQR)]
df_iqr
```

# Z-score method

```python
from scipy.stats import zscore
```

```python
df['z_score'] = zscore(df['beer_servings'])
```

```python
df_zscore = df[(df['z_score'] < 3) & (df['z_score'] > -3)]
df_zscore.drop(columns=['z_score'])
df_zscore
```
